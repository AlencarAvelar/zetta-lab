```python
import pandas as pd 

# Caminhos
shapefile_path = "/home/alencaravelar/Desktop/zetta-lab/zetta-lab/data/raw/shapefiles/RM_BeloHorizonte_UDH_region.shp"
csv_path = "/home/alencaravelar/Desktop/zetta-lab/zetta-lab/data/refined/base_udh_refined.csv"

# Ler os dados
geo = gpd.read_file(shapefile_path)
df = pd.read_csv(csv_path)

# Garantir tipos de string
geo["UDH_ATLAS"] = geo["UDH_ATLAS"].astype(str)
df["Cod_ID"] = df["Cod_ID"].astype(str)


df["UDH_ATLAS"] = df["Cod_ID"]

# Fazer o merge
geo_merged = geo.merge(df, on="UDH_ATLAS", how="inner")

# Verificar correspondência
print("Linhas no shapefile:", len(geo))
print("Linhas no CSV:", len(df))
print("Linhas após merge:", len(geo_merged))

# Plotar o mapa
if len(geo_merged) > 0:
    plt.figure(figsize=(10, 10))
    geo_merged.plot(
        column="IDHM",
        cmap="YlGn",
        legend=True,
        legend_kwds={'label': "Índice de Desenvolvimento Humano (IDHM)"},
        linewidth=0.3,
        edgecolor="gray"
    )
    plt.title("🌎 Mapa da Região Metropolitana de Belo Horizonte - IDHM (2010)")
    plt.axis("off")
    plt.show()
else:
    print("⚠️ Nenhuma correspondência encontrada — verifique o código municipal extraído.")
```
